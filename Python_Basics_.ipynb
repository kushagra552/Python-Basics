 - The == and is operators in Python serve distinct comparison purposes:
== (Equality): It checks if the values of two objects are the same. It compares the content or data held by the objects.
is (Identity): It verifies if two variables refer to the exact same object in memory. It checks if the memory addresses of the objects are identical.

10. What is the difference between implicit and explicit type casting

13. What is the difference between for and while loops
 - For loop is used to iterate over a sequence of items. While loop is used to repeatedly execute a block of statements while a condition is true. For loops are designed for iterating over a sequence of items.
14. Describe a scenario where a while loop is more suitable than a for loop
 - A while loop is more suitable than a for loop when you need to repeat a block of code until a specific condition is met, where the exact number of iterations is unknown beforehand, such as when prompting a user for valid input until they provide the correct format, as you don't know how many times they might need to try before getting it right; in this case, the loop would continue until the input is valid, regardless of how many attempts it takes.

```python
# 1. Write a Python program to print "Hello, World!

print("Hello, World!")

# 2. Write a Python program that displays your name and age

my_name = "mohit"
my_age = 22

print(f"My name is {my_name} and I am {my_age} years old.")

# 3. Write code to print all the pre-defined keywords in Python using the keyword library

import keyword

print(keyword.kwlist)

# 4. Write a program that checks if a given word is a Python keyword

import keyword

def is_keyword(word):
  """Checks if a given word is a Python keyword.

  Args:
    word: The word to check.

  Returns:
    True if the word is a keyword, False otherwise.
  """
  return word in keyword.kwlist

word_to_check = input("Enter a word: ")

if is_keyword(word_to_check):
  print(f"'{word_to_check}' is a Python keyword.")
else:
  print(f"'{word_to_check}' is not a Python keyword.")

# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differntly for each.

my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)
my_list[2] = 10
print("Modified list:", my_list)
print("Tuple:", my_tuple)

# 6. Write a function to demonstrate the behavior of mutable and immutable arguments

def modify_arguments(number, my_list):
  """Demonstrates the behavior of mutable and immutable arguments.

  Args:
    number: An integer (immutable).
    my_list: A list (mutable).
  """
  number += 10
  my_list.append(10)
  print("Inside the function:")
  print("number:", number)
  print("my_list:", my_list)

# 7. Write a function to demonstrate the behavior of mutable and immutable arguments

def modify_arguments(number, my_list):
    """Demonstrates the behavior of mutable and immutable arguments.

    Args:
        number: An integer (immutable).
        my_list: A list (mutable).
    """

print("Inside the function, before modification:")
print("number:", number, "id(number):", id(number))
print("my_list:", my_list, "id(my_list):", id(my_list))
number += 10
my_list.append(10)
print("\nInside the function, after modification:")
print("number:", number, "id(number):", id(number))
print("my_list:", my_list, "id(my_list):", id(my_list))
original_number = 5
original_list = [1, 2, 3]

print("Outside the function, before calling:")
print("original_number:", original_number, "id(original_number):", id(original_number))
print("original_list:", original_list, "id(original_list):", id(original_list))

modify_arguments(original_number, original_list)

print("\nOutside the function, after calling:")
print("original_number:", original_number, "id(original_number):", id(original_number))
print("original_list:", original_list, "id(original_list):", id(original_list))

# 8. Write a program to demonstrate the use of logical operators

age = int(input("Enter your age: "))
has_license = input("Do you have a driver's license? (yes/no): ").lower()
if age >= 18 and has_license == "yes":
print("You are eligible to drive.")
elif age < 18:
print("You are too young to drive.")
elif has_license != "yes":
print("You need a driver's license to drive.")
is_weekend = input("Is it a weekend? (yes/no): ").lower()
is_holiday = input("Is it a holiday? (yes/no): ").lower()
if is_weekend == "yes" or is_holiday == "yes":
print("You can relax and enjoy!")
else:
print("It's a working day.")
is_raining = input("Is it raining? (yes/no): ").lower()
if not is_raining == "yes":
print("You can go outside.")
else:
print("Stay indoors, it's raining.")

# 9. Write a Python program to convert user input from string to integer, float, and boolean types

def convert_input_types():
    user_input = input("Enter a value: ")

    try:
        integer_value = int(user_input)
        print(f"Integer: {integer_value}")
    except ValueError:
        print("Not an integer.")

    try:
        float_value = float(user_input)
        print(f"Float: {float_value}")
    except ValueError:
         print("Not a float.")

    if user_input.lower() in ["true", "false"]:
        boolean_value = user_input.lower() == "true"
        print(f"Boolean: {boolean_value}")
    else:
        print("Not a boolean.")

convert_input_types()

# 10. Write code to demonstrate type casting with list elements

my_list = [10, "20", 30.5, "40", True]
integers_list = [int(x) for x in my_list if str(x).isdigit()]
print("Integers:", integers_list)
floats_list = [float(x) for x in my_list if isinstance(x, (int, float, str)) and str(x).replace('.', '', 1).isdigit()]
print("Floats:", floats_list)
strings_list = [str(x) for x in my_list]
print("Strings:", strings_list)
booleans_list = [bool(x) for x in my_list]
print("Booleans:", booleans_list)


# 11. Write a program that checks if a number is positive, negative, or zero

def check_number(number):
    """Checks if a number is positive, negative, or zero.

    Args:
        number: The number to check.

    Returns:
        A string indicating whether the number is positive, negative, or zero.
    """
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"
number = float(input("Enter a number: "))
result = check_number(number)
print(f"The number is {result}.")

# 12. Write a for loop to print numbers from 1 to 100

for i in range(1, 101):
    print(i)

# 13. Write a Python program to find the sum of all even numbers between 1 and 500

sum_even = 0
for number in range(2, 501, 2):
    sum_even += number
print(f"The sum of even numbers between 1 and 500 is: {sum_even}")

# 14. Write a program to reverse a string using a while loop

def reverse_string(text):
    """Reverses a string using a while loop.

    Args:
        text: The string to reverse.

    Returns:
        The reversed string.
    """
    reversed_text = ""
    index = len(text) - 1

    while index >= 0:
        reversed_text += text[index]
        index -= 1

    return reversed_text

input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)


# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

def factorial_while_loop(num):
    """
    Calculate the factorial of a number using a while loop.
    """
    if num < 0:
        return "Factorial is not defined for negative numbers."
    elif num == 0:
        return 1
    else:
        factorial = 1
        while num > 0:
            factorial *= num
            num -= 1
        return factorial
try:
    number = int(input("Enter a non-negative integer: "))
    result = factorial_while_loop(number)
    print(f"The factorial of {number} is {result}")
except ValueError:
    print("Invalid input. Please enter a non-negative integer.")












```
